## Predicting the Optimum Number of Clusters using Python's KMeans Clustering

```python
#Import all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets

from sklearn.cluster import KMeans
```

```python
#Load the dataset
iris=datasets.load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data.head(4)
```

### Brief Introduction to Clustering

Clustering is the process of dividing data into groups (clusters) based on patterns in the data. It is an unsupervised machine learning because there are target variables, our model looks at the data and tries to group them based on patterns in the data. Data points in a cluster are similar to one another and different from data points in different clusters.
**It is important to know the optimum number of clusters (K) when using K Means clustering. We'll do this by using the Elbow Method**

### Finding the Optimum number of Clusters using the Elbow Method

```python
distortions = []
cluster = range(1,11)
for i in cluster:
    Model = KMeans(n_clusters=i)
    Model.fit(data)
    distortions.append(Model.inertia_)
#Inertia is the sum of squared error for each cluster, the smaller the inertia the denser the cluster 

#Visualizing the Elbow Method
plt.figure(figsize=(12,8))
plt.plot(cluster,distortions)
plt.show()
```

**Based on our Elbow Point method, the value for our n_cluster should be 3**

## Visualizing our clusters

```python
kmeans=KMeans(n_clusters=3)
y=kmeans.fit_predict(data)
x=data.values

plt.figure(figsize=(10,5))
plt.scatter(x[y == 0, 0], x[y == 0, 1], 
             c = 'blue', label = 'Iris-setosa')
plt.scatter(x[y == 1, 0], x[y == 1, 1], 
              c = 'red', label = 'Iris-versicolour')
plt.scatter(x[y == 2, 0], x[y == 2, 1],
              c = 'black', label = 'Iris-virginica')

plt.legend()
plt.show()
```
